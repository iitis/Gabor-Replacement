```python
# Functions used to generate a Gabor filter bank and to modify a pretrained CNN model from keras
from Gabor_replacement_CNNs import generate_Gabor_filter_bank, replace_CNN_filters_with_Gabor, generate_matching_random_filter_bank

# Pretrained model from keras and its preprocessing function
from tensorflow.keras.applications import Xception
from tensorflow.keras.applications.inception_v3 import preprocess_input

# Helper function that 'translates' onehot prediction to human-readable form
from tensorflow.keras.applications.imagenet_utils import decode_predictions

# Reading images, processing them
from skimage.io import imread
from skimage.transform import resize
import cv2

# For array operations
import numpy as np

# Visualization of the image
import matplotlib.pyplot as plt
```

First, we load one of the pretrained models from Keras. We load 2 models to two variables. One of them will be modified with the random filters.

```python
# In our demo, we use one of the pretrained models from Keras Applications - Xception 
# Reference: Xception: Deep Learning with Depthwise Separable Convolutions (CVPR 2017)

# model to be mofdified:
model = Xception(
    include_top=True,
    weights="imagenet",
    input_tensor=None,
    input_shape=None,
    pooling=None,
    classes=1000,
    classifier_activation="softmax",
)

# model to be kept untouched
model0 = Xception(
    include_top=True,
    weights="imagenet",
    input_tensor=None,
    input_shape=None,
    pooling=None,
    classes=1000,
    classifier_activation="softmax",
)
```

We use our function (**generate_Gabor_filter_bank**) to generate the filter bank:

```python
gabor_filter_bank = generate_Gabor_filter_bank(filter_size=3, period_min=1, period_max=10, 
                                               deg_min=0, deg_max=180, deg_step=5, phase_min=-5, 
                                               phase_max=5, phase_step=0.3, gaussian_std_min=0.1, 
                                               gaussian_std_max=30, gaussian_std_step=0.2)
```

Generate a matching random filter bank

```python
random_filter_bank = generate_matching_random_filter_bank(gabor_filter_bank)
```

Now, we can use our function (**replace_CNN_filters_with_Gabor**) to replace all the first layer filters with the most similar filters from our filter bank according to the cosine similarity.

```python
model = replace_CNN_filters_with_Gabor(model, random_filter_bank, 1)
```

We use the imported functions to read and preprocess the image. We use dimension 299x299, as this is the required dimension from the Xception model. We also use Xception's preprocessing function from Keras, that scales the uinput data to range -1, 1

```python
dark_image = imread('indian_elephant.jpg')
image = cv2.resize(dark_image, (299, 299), interpolation = cv2.INTER_NEAREST)
image = preprocess_input(image)
image = np.expand_dims(image, axis=0)


plt.imshow(image[0] * 0.5 + 0.5)
plt.show()
```

Now, we can generate the prediction of the original model - model0:

```python
decode_predictions(model0.predict(image))
```

We can see, that the prediction changes after the filter replacement. 

```python
decode_predictions(model.predict(image))
```
